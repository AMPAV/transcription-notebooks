# Overview of AMPAV

Architectually, AMPAV is built in several layers, each using functionality
from layer below.  From bottom to top, the structure is:
* The core (`ampav.core`) provides
    * Commonly used utility functions like duration to human-readable time, etc
    * Schemas which are python classes representing different types of metadata
      that can be manipulated in memory or stored if needed
* Tools (such as `ampav.whisper`) which extract metadata and return 
  different schema objects (from `ampav.core`) using media files or other
  resources.  
* Pipelines the tools (perhaps in combination) to perform metadata augmentation:
    * Jupyter notebooks (such as these) can be used for experimentation and
      evaluation of tools on your local computer
    * Batch processing scripts can be run on a server to process mass quanities
      of content
    * REST services can be built to provide augmentation services that can be
      used by other software, such as a transcription service
  
Let's begin with a short tour of `ampav.core`

## duration2hhmmss and hhmmss2seconds

The `duration2hhmmss` function takes a duration in seconds and produces a 
human-readable representation of it. 

The first step is to import the function into our code

```python
from ampav.core.utils import duration2hhmmss
```

Now that it's in our code we can use it.  Let's say we have something that is
`3661.001` seconds long.  The function should return 1 hour, 1 minute, 1 second,
and 1 1/1000ths of a second

```python
duration = 3661.001
print(duration2hhmmss(duration))
```

Feel free to change the value of duration to different times and see how the 
output changes.  Remember that an hour is 3600 seconds and a minute is 60 
seconds.

## AVMetadata
The `AVMetadata` schema class can be used to extract technical metadata from a 
AV media file.  Things like the duration, number of audio streams, how the
media was encoded, etc, are available using this object.

As before, the first thing to do is to import it into our code:

```python
from ampav.core.schema.av_metadata import AVMetadata
```

Now that is is loaded, let's get more information about the `gettysburg.wav` 
file that's in our project.  We'll store that information in the `avm` 
variable and print out some of the details

```python
filename = "gettysburg.wav"
avm = AVMetadata.from_file(filename)
print(f"Format {avm.format_name}")
print(f"Duration: {avm.duration}")
print(f"Number of audio streams: {len(avm.streams.audio)}")
print(f"Number of video streams: {len(avm.streams.video)}")
```

The duration is a little bit hard to read -- using math we can determine that
it's 3 minutes, 3.04 seconds long.  But math is hard, so let's print it again 
using the `duration2hhmmss` function

```python
print(f"Duration: {duration2hhmmss(avm.duration)}")
```

If you want to see all of the properties of the file you can simply print the
variable

```python
print(avm)
```

All schema objects in AMPAV can also be stored for later use.  One format that
they can be stored in is called YAML which can be read by both computers and
humans.  Let's print the YAML representation

```python
print(avm.model_dump_yaml())
```
