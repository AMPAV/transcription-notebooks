# Nvidia Parakeet

Nvidia's NeMo platform provides a speech to text model called parakeet.  AMPAV
provides a wrapper around this model to make it easy to use

NOTE: You may need to restart the kernel for this notebook to make sure enough
memory is free to run the code blocks.

As before, let's import the tool function into the code.  We'll also import the
`Path` class

```python
from ampav.parakeet.transcribe import transcribe_file
from pathlib import Path
```

The `transcribe_file` function takes several arguments (which are different
from the Whisper ones):
* A `Path` object which is the file to transcribe
* The name of the model to use (one of)
    * `nvidia/parakeet-tdt-0.6b-v2`
    * `nvidia/parakeet-tdt-0.6b-v3`
* A True/False value for forcing it to CPU only
* Audio chunk size - process the audio in chunks rather than all at once
* Audio overlap size - how much overlap each audio chunk should have with
  the next one

As in Whisper, when changing the model or device you may need to restart the 
kernel and re-run all of the code blocks.  If you set `cpu_only` to False, 
you may have to modify the `chunk_size` and `chunk_overlap` to fit your GPU.


```python
file = Path("gettysburg.wav")
modelname = "nvidia/parakeet-tdt-0.6b-v3"
cpu_only = True
chunk_size = 30
chunk_overlap = 5
ts = transcribe_file(file, modelname, cpu_only, chunk_size, chunk_overlap)
```

Parakeet's transcription log output is very chatty and you can ignore any
information or warning messages.

As in the Transcripts notebook, `transcribe_file` has returned a `Transcript`
object and we can view the same information about the tool and the transcript
itself

```python
print(f"Tool name: {ts.tool_name}")
print(f"Tool version: {ts.tool_version}")
print(f"Tool run time: {ts.run_duration()}")
print(f"First 30 characters of the transcript: {ts.output.text[0:30]}")
print(f"Number of words in the transcript: {len(ts.output.words)}")
print(f"Number of paragraphs in the transcript: {(len(ts.output.paragraphs))}")
```

This time, we can see that the tool's runtime is populated because this tool
provides this information.  It also provides more information about the 
parameters used by the tool

```python
print(f"Tool parameters: {ts.parameters}")
```

We can view all of the transcript details by printing the YAML version of the
object

```python
print(ts.model_dump_yaml())
```

As in the Whisper tutorial we can generate a WebVTT

```python
from ampav.core.file_formats.webvtt import paragraphs_to_webvtt
print(paragraphs_to_webvtt(ts.output.paragraphs))
```
