# Installing the AMPAV packages

AMPAV is distributed as a set of python packages.  

The `ampav.core` package provides the core functionality that is used by all of
the other packages.  Other packages may provide tooling which to speech to text,
named entity recognition, scene detection, or other tools.  

To install the current ampav packages, run the code block below (it may take a
long time but it only needs to be run once)

````python
%pip install -U --extra-index-url https://pypi.dlib.indiana.edu ampav-core==0.0.12 ampav-whisper==0.0.12 ampav-parakeet==0.0.12
````

If the installation is successful, the output should end with something like:
```
Successfully installed <lots of packages across multiple lines>
Note: you may need to restart the kernel to use the updated packages
```

After a successful install the code below should run and print out 
"Hello AMPAV".  There may be some warnings from the underlying libraries, 
but they can be ignored.  Loading the transcribe packages may take a while.

````python
import ampav.core.media
import ampav.whisper.transcribe
import ampav.parakeet.transcribe
print("Hello AMPAV")
````
