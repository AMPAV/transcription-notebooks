# Transcripts
`ampav.core` provides a schema for storing and manipulating transcript data.  
It doesn't really matter how the transcript was created it could have come from
a file, a local tool, or something from the cloud.  In every case the transcript
objects are all identical.

To start, let's import the `Transcript` schema into our code.  We would also
like to load a text file and convert it into a transcript, so we will need to
import `import_text_transcript` as well

```python
from ampav.core.schema.transcript import Transcript
from ampav.core.file_formats.textfile import import_text_transcript
```

A transcript consists of several parts:
* The raw text
* A list of words that were found in the text 
* A list of paragraphs that were found

Words and paragraphs can have start/end timing information, if the tool that
created it provides it.

The `import_text_transcript` is an example of a core function that acts like
a tool.  It will take a string and produce a `ToolOutput`, which is a
wrapper schema that wraps the actual output (in this case, a `Transcript`) and
also provides metadata about the tool that created it, such as the name,
version, and start/end times.

Let's create a transcript from the text in the `gettysburg.txt` file

```python
filename = "gettysburg.txt"
with open(filename) as f:
    ts = import_text_transcript(f.read())
```

The warning appears because this is a simple text file -- the words do not 
have timing information so it is going to estimate where the paragraphs are
split.

Because this function acts like a tool, the result is a `ToolObject` which 
gives us some information.

```python
print(f"Tool name: {ts.tool_name}")
print(f"Tool version: {ts.tool_version}")
print(f"Tool run time: {ts.run_duration()}")
```

The run time is `None` because the tool didn't provide any start/end times. 
Other transcript tools would likely do so.

The output that the tool created is in the `output` property of the tool.  Let's
print some information about the transcript

```python
print(f"First 30 characters of the transcript: {ts.output.text[0:30]}")
print(f"Number of words in the transcript: {len(ts.output.words)}")
print(f"Number of paragraphs in the transcript: {(len(ts.output.paragraphs))}")
```

As with the `AVMetadata` schema, we can print a YAML rendering of the output

```python
print(ts.model_dump_yaml())
```
